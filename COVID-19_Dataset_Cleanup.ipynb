# List of Cities with FIPS region code

```python
df_cities = pd.read_csv('data/city_names.csv')
```

```python
df_cities.head()
```

```python
df_cities=df_cities.rename(columns={"State Name": "state", "City Name": "city", "County Name": "county", "State Code": "state code", "City Code": "city code", "County Code": "county code"})
```

Adding region codes to Colleges.csv

```python
df_college['state_lower'] = df_college['state'].str.lower()
df_college['county_lower'] = df_college['county'].str.lower()
df_college['city_lower'] = df_college['city'].str.lower()
df_college.head()
```

```python
df_cities['state_lower'] = df_cities['state'].str.lower()
df_cities['county_lower'] = df_cities['county'].str.lower()
df_cities['city_lower'] = df_cities['city'].str.lower()
df_cities.head()
```

```python
result = pd.merge(df_college, df_cities, how='left', on=['state_lower', 'county_lower','city_lower'])
result.head(10)
```

```python
result=result.drop(['state_lower', 'county_lower', 'city_lower', 'state_y', 'county_y', 'city_y'], axis=1)
```

```python
result.shape
```

```python
#df=result
#nan_values = df[df['city code'].isnull()]
#nan_values
result = result.dropna()
```

```python
result.shape
```

```python
result['ipeds_id'] = result['ipeds_id'].astype(int)
result['cases'] = result['cases'].astype(int)
result['city code'] = result['city code'].astype(int)
result.dtypes
```

```python
result=result.rename(columns={"state_x": "state", "city_x": "city", "county_x": "county", "state code": "state_code", "city code": "city_code", "county code": "county_code"})
```

```python
result.head()
```

```python
result = result[['date', 'state', 'state_code', 'county', 'county_code', 'city', 'city_code', 'ipeds_id', 'college', 'cases']]
df_college=result
df_college.head()
```

```python
df_cities=df_cities.drop(['state_lower', 'county_lower', 'city_lower'], axis=1)
```

```python
df_cities=df_cities.rename(columns={"state code": "state_code", "city code": "city_code", "county code": "county_code"})
```

# Masks

```python
df_masks = pd.read_csv('data/mask-use-by-county.csv')
df_masks.head()
```

```python
df_masks.isnull().sum()
```

```python
df_masks.describe()
```

```python
df_masks = df_masks.rename(columns={"COUNTYFP":"fips"})
df_masks.head()
```

```python
df_masks1 = df_counties.merge(df_masks, on='fips')
df_masks1.head()
```
